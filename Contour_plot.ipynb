```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as col
```

```python
arr = np.array([100,110,15,15,14,12,100,12,12,9,12,12,5,5,4,3,5,5,5,3,4,3,1,2,3,5,4,4,7,5,4,5,4,6,3,100])
arr = arr.reshape([6,6]) # 2D
x = np.arange(0,36,6)
y=np.arange(0,36,6)
plt.get_cmap('jet') #viridis could work too
```

```python
arr = arr[::-1] #flip the 2D array so it plots as a book is read
```

```python
fig, ax = plt.subplots()
c = ax.contourf(x,y,arr,levels=1000,cmap='jet')
plt.colorbar(c,label = 'Temperature ($ \degree$C)')
ax.set_xlabel('$x$-Position (cm)')
ax.set_ylabel('$y$-Position (cm)');
```
