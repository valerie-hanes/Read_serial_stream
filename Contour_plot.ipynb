```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as col
```

```python
arr = np.array([12,11,5,5,4,12,13,12,12,9,12,12,30,21,17,16,25,24,23,3,4,3,1,2,3,8,8,9,12,23,33,5,4,6,3,4])
arr = arr.reshape([6,6])
x = np.arange(0,36,6)
y=np.arange(0,36,6)
plt.get_cmap('jet')
```

```python
fig, ax = plt.subplots()
c = ax.contourf(x,y,arr,levels=1000,cmap='viridis')
plt.colorbar(c,label = 'Temperature ($ \degree$C)')
ax.set_xlabel('$x$-Position (cm)')
ax.set_ylabel('$y$-Position (cm)');
```
